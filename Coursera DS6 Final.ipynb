```python
%matplotlib inline

import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm 
import scipy.stats as stats
from sklearn import preprocessing
from scipy.stats import normaltest
```

```python
df = pd.read_csv('kc_house_data.csv')
df.head()
```

```python
df.dtypes
```

```python
drop_list = ["id"]
df = df.drop(drop_list,axis=1)
df.describe()
```

```python
df.floors.value_counts().to_frame()
```

```python
df.boxplot(by = 'waterfront',column = ["price"], grid = False)
```

```python
#import seaborn as sns, sns.set(color_codes=True)
#price = sns.load_dataset("price")
sns.regplot(x=df["sqft_above"], y=df["price"], data=df["price"])
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
lm=LinearRegression()
X = df[['floors','waterfront','lat','bedrooms','sqft_basement','view',\
        'bathrooms','sqft_living15','sqft_above','grade','sqft_living']]
Y = df['price']
lm.fit(X, Y)
Y_pred=lm.predict(X)
r2_score(Y, Y_pred)
```

```python
#pr=PolynomialFeatures(degree=2)
#pr.fit_transform(X, include_bias=False)
#print(p)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X = df[['floors','waterfront','lat','bedrooms','sqft_basement','view','bathrooms','sqft_living15']]
Y = df['price']

Input=[('scale',StandardScaler()),('polynomial',PolynomialFeatures(degree=8)),('mode',LinearRegression())]

pipe = Pipeline(Input)
pipe.fit(X,Y)
y_pred = pipe.predict(X)
r2_score(Y,y_pred)


```

```python
from sklearn.linear_model import Ridge
RigeModel = Ridge(alpha=0.1)
X = df[['floors','waterfront','lat','bedrooms','sqft_basement','view',\
        'bathrooms','sqft_living15','sqft_above','grade','sqft_living']]
Y = df['price']
RigeModel.fit(X,Y)
Y_pred = RigeModel.predict(X)
r2_score(Y, Y_pred)
```
